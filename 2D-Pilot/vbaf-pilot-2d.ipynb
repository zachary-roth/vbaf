## (A) Create a New Project

```python
deeplabcut.create_project.new.create_new_project(project = 'vbaf-pilot-2d',
                                                 experimenter = 'zr-bvdb', 
                                                 videos = path_videos, 
                                                 working_directory='/home/zach/Documents/vbaf-pilot-2d/', 
                                                 copy_videos=True, 
                                                 videotype='.mp4')
```

## (B) Configure the Project

- Change **body parts** to list of markers
- Change **colormap** to spring
- Change **dotsize** to 2
- Change **default_net_type** to resnet_152
- Change **default_augmenter** to imgaug
- Change **batch_size** to 4

```python
config_path = '/home/zach/Documents/vbaf-pilot-2d/vbaf-pilot-2d-zr-bvdb-2022-08-25/config.yaml'
```

## (C) Data Selection (extract frames)

```python
deeplabcut.extract_frames(config_path, mode='automatic', algo='uniform', crop=True, userfeedback = False)
```

## (D) Label Frames

```python
deeplabcut.label_frames(config_path)
```

## (E) Check Annotated Frames

```python
deeplabcut.check_labels(config_path, visualizeindividuals=False)
```

## (F) Create Training Dataset(s)

```python
deeplabcut.create_training_dataset(config_path, num_shuffles=1, net_type='resnet_152', augmenter_type='imgaug')
```

## (G) Train The Network

```python
deeplabcut.train_network(config_path, shuffle=1, trainingsetindex=0, 
                         max_snapshots_to_keep=10, displayiters=500, 
                         saveiters=10000, maxiters=300000, allow_growth=True)
```

## (H) Evaluate the Trained Network

```python
deeplabcut.evaluate_network(config_path,Shuffles=[1], plotting=True,show_errors=True,rescale=True)
```
